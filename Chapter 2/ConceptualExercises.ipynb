**5** The advantages of using a flexible versus a less flexible approach when using regression/classification is that when your using a flexible model the interpretability of the model decreases the more flexible the model. When using a less flexible approach, the interpretability of the model increases the higher the less flexibe the model. Cases where you would prefer a flexible approach over a less flexible approach is if the interpretability of the model does not matter and you are mainly focued on the results or predictions. Cases where you would prefer a less flexible model over a flexible model is if you are mainly focused on inference or the interpretability of the model.

**6** The differences between a parametric approach and a non-parametric approach is that a parametric approach makes an assumption about the true form of f and a non-parametric approach does not make assumptions about the true form of f. The advantages of using a parametric approach is that you can simplify the model by estimating parameters with fewer observations opposed to a non-parametric function that attempt to fit a arbitrary function f with many observations. Disadvantage in using a parametric approach is you could obtain the true functional form f and overfit the data. Disadvantages in using a non-parametric approach is that it does reduce the problem of estimating parameters and it requires a very large number of observations to get the true functional form of f.
