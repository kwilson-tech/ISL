```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 8(a)

college = pd.read_csv('../Files/College.csv')
```

```python
# 8(b) 

college
```

```python
#Removed the first column
college = college.drop('Unnamed: 0',axis=1)

college
```

```python
# 8(c)i. 

college.describe().transpose()
```

```python
# 8(c)ii. 

# Private column is categorical and this function only plots numerical
# so I added the hue parameter for Private feature. 
sns.pairplot(data=college.iloc[:,0:10],hue='Private')
plt.show()
```

```python
# 8(c)iii. 

plt.figure(figsize=(5,5),dpi=150)
sns.boxplot(data=college,x='Private',y='Outstate')
plt.show()
```

```python
# 8(c)iv. 

college['Elite'] = college['Top10perc'] > 50
```

```python
college['Elite'].value_counts()
```

```python
plt.figure(figsize=(5,5),dpi=150)
sns.boxplot(data=college,x='Elite',y='Outstate');
```

```python
# 8(c)v. 

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(5,5),dpi=150)

axs[0,0].hist(data=college,x='Outstate',bins=30)
axs[0,1].hist(data=college,x='Enroll',bins=30)
axs[1,0].hist(data=college,x='Room.Board',bins=30)
axs[1,1].hist(data=college,x='Books',bins=30)

axs[0,0].set_title('Outstate')
axs[0,1].set_title('Enroll')
axs[1,0].set_title('Room Board')
axs[1,1].set_title('Books')

plt.tight_layout()
plt.show()
```

```python
# 8(c)vi. Continue exploring the data, and provide a brief summary
# of what you discover.

fig, axs = plt.subplots(nrows=1, ncols=2,figsize=(6,3),dpi=200)

sns.scatterplot(ax=axs[0],data=college,x='Apps',y='Accept',hue='Private')
sns.scatterplot(ax=axs[1],data=college,x='Apps',y='Outstate',hue='Private')

axs[0].set_title('Accept vs Apps')
axs[1].set_title('Outstate vs Apps')

plt.tight_layout()
plt.show()
```

**8(c)vi.**

I was curious about the number of applicants that applied to private institutions, so 
we see that a lot of private institutions received applications from students out of state. 
This coincides with the Accept vs Apps graph that shows that Private institutions acceptance rate
is quite low compared to public colleges.
