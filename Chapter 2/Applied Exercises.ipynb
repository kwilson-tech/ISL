```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 8(a) Use the read.csv() function to read the data into R. Call the loaded data college...

college = pd.read_csv('../CSV Files/College.csv')
```

```python
# 8(b) Look at the data using the View() function. You should notice
# that the first column is just the name of each university...

college
```

```python
#Removed the first column
college = college.drop('Unnamed: 0',axis=1)

college
```

```python
# 8(c) i. Use the summary() function to produce a numerical summary
# of the variables in the data set...
college.describe().transpose()
```

```python
# 8(c) ii. Use the pairs() function to produce a scatterplot matrix of
# the first ten columns or variables of the data...

# Private column is categorical and this function only plots numerical
# so I added the hue parameter for Private feature (double click on plot to zoom in). 
sns.pairplot(data=college.iloc[:,0:10],hue='Private');
```

```python
# 8(c) iii. Use the plot() function to produce side-by-side boxplots of
# Outstate versus Private...

plt.figure(figsize=(5,5),dpi=150)
sns.boxplot(data=college,x='Private',y='Outstate');
```

```python
# 8(c) iv. Create a new qualitative variable, called Elite, by binning
# the Top10perc variable...

college['Elite'] = college['Top10perc'] > 50
```

```python
college['Elite'].value_counts()
```

```python
plt.figure(figsize=(5,5),dpi=150)
sns.boxplot(data=college,x='Elite',y='Outstate');
```

```python
# 8(c) v. Use the hist() function to produce some histograms with
# differing numbers of bins for a few of the quantitative variables...

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(5,5),dpi=150)

axs[0][0].hist(data=college,x='Outstate',bins=30)
axs[0][1].hist(data=college,x='Enroll',bins=30)
axs[1][0].hist(data=college,x='Room.Board',bins=30)
axs[1][1].hist(data=college,x='Books',bins=30)

axs[0][0].set_title('Outstate')
axs[0][1].set_title('Enroll')
axs[1][0].set_title('Room Board')
axs[1][1].set_title('Books')

plt.tight_layout()
plt.show()
```
